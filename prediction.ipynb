```python
import tensorflow as tf
import pandas as pd
import numpy as np
import pickle
from tensorflow.keras.models import load_model
```

```python
## load the trained model, scaler pickle and onehot
model=load_model('model.h5')

##load the encoder and scaler pickle files
with open('one_hot_encoder_geo.pkl','rb') as file:
    one_hot_encoder_geo=pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
## example input data for prediction
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 2,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
## one hot encode the 'Geography' column
geo_encoded=one_hot_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
print(type(label_encoder_gender))
print(label_encoder_gender.classes_)
print(input_df['Gender'])
```

```python
##encode categorical variables
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
## contatination one hot encoded
input_df=pd.concat([input_df.drop("Geography",axis=1),geo_encoded_df],axis=1)
input_df
```

```python
## scaling the input data
scaled_input=scaler.transform(input_df)
scaled_input
```

```python
## prediction
prediction=model.predict(scaled_input)
prediction
```

```python
prediction_prob=prediction[0][0]
```

```python
prediction_prob
```

```python
if prediction_prob > 0.5:
    print("The customer is likely to churn.")
else:
    print("The customer is not likely to churn.")
```
